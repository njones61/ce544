```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kr = 0.005
ks = 0.001
alpha = -60
```

```python
# define functions

def kxx(kr, ks, alpha):
  a_rad = alpha * np.pi/180 # convert degrees to radians
  return kr * np.cos(a_rad)**2 + ks * np.sin(a_rad)**2

def kyy(kr, ks, alpha):
  a_rad = alpha * np.pi/180
  return kr * np.sin(a_rad)**2 + ks * np.cos(a_rad)**2

def kxy(kr, ks, alpha):
  a_rad = alpha * np.pi/180
  return - (1/2) * (kr - ks) * np.sin(2*a_rad)
```

```python
k_xx = kxx(kr, ks, alpha)
k_yy = kyy(kr, ks, alpha)
k_xy = kxy(kr, ks, alpha)

print(f'k_xx = {k_xx:.4f}')
print(f'k_yy = {k_yy:.4f}')
print(f'k_xy = {k_xy:.4f}')
```

```python
# Create a numpy array ranging from 0 to -90 by increments of 5
angles = np.arange(0, -95, -5)
```

```python
kx = kxx(kr, ks, angles)
ky = kyy(kr, ks, angles)
```

```python
# plot kx and ky with a legend and axis titles
plt.figure(figsize=(10, 4))
plt.plot(angles, kx, label='kxx')
plt.plot(angles, ky, label='kyy')
plt.xlabel('Angle (degrees)')
plt.ylabel('Hydraulic Conductivity')
plt.title('Hydraulic Conductivity vs Angle')
plt.legend()
plt.show()
```
